```python
import matplotlib.pyplot as plt
import networkx as nx

# Define the neural network architecture
layers = {
    "Input Layer": 4,
    "Hidden Layer 1": 3,
    "Hidden Layer 2": 2,
    "Output Layer": 1
}

# Create the network graph
def plot_neural_network(layers):
    G = nx.DiGraph()
    layer_positions = {}
    node_positions = {}

    current_x = 0
    node_id = 0

    # Assign nodes and positions for each layer
    for layer, num_nodes in layers.items():
        layer_positions[layer] = current_x
        for i in range(num_nodes):
            node_positions[node_id] = (current_x, -i)
            G.add_node(node_id, layer=layer)
            node_id += 1
        current_x += 1

    # Connect the nodes between layers
    for layer_idx, (layer, num_nodes) in enumerate(layers.items()):
        if layer_idx == len(layers) - 1:
            break
        current_layer_nodes = [
            node_id for node_id, attrs in G.nodes(data=True) if attrs["layer"] == layer
        ]
        next_layer = list(layers.keys())[layer_idx + 1]
        next_layer_nodes = [
            node_id for node_id, attrs in G.nodes(data=True) if attrs["layer"] == next_layer
        ]
        for current_node in current_layer_nodes:
            for next_node in next_layer_nodes:
                G.add_edge(current_node, next_node)

    # Plot the graph
    plt.figure(figsize=(10, 6))
    nx.draw(
        G,
        pos=node_positions,
        with_labels=False,
        node_size=1000,
        node_color="lightblue",
        edge_color="gray",
        width=1.5
    )
    # Add layer labels
    for layer, x_pos in layer_positions.items():
        plt.text(x_pos, 1, layer, fontsize=12, ha="center", va="bottom", fontweight="bold")
    plt.title("Neural Network Architecture", fontsize=14)
    plt.axis("off")
    plt.show()

# Plot the network
plot_neural_network(layers)
```

```python
from graphviz import Digraph

# Initialize Digraph
nn_graph = Digraph(format='png')
nn_graph.attr(rankdir='LR', splines='line', nodesep='0.8', ranksep='0.8')

# Set uniform node attributes
nn_graph.attr('node', shape='circle', style='filled', fixedsize='true', width='0.8', height='0.8')

# Input layer
with nn_graph.subgraph(name='cluster_input') as input_layer:
    input_layer.attr(label="Input Layer", color='black', style='dashed')
    inputs = ['X1', 'X2', 'X3', 'X4']
    for i in inputs:
        input_layer.node(i, color='lightyellow', fontcolor='black')

# Hidden layer 1
with nn_graph.subgraph(name='cluster_hidden1') as hidden1_layer:
    hidden1_layer.attr(label="Hidden Layer L1 (tanh)", color='blue', style='dashed')
    hidden1 = ['A1_1', 'A1_2', 'A1_3']
    for h1 in hidden1:
        hidden1_layer.node(h1, color='skyblue', fontcolor='black')

# Hidden layer 2
with nn_graph.subgraph(name='cluster_hidden2') as hidden2_layer:
    hidden2_layer.attr(label="Hidden Layer L2 (tanh)", color='green', style='dashed')
    hidden2 = ['A2_1', 'A2_2']
    for h2 in hidden2:
        hidden2_layer.node(h2, color='skyblue', fontcolor='black')

# Output layer
with nn_graph.subgraph(name='cluster_output') as output_layer:
    output_layer.attr(label="Output Layer (sigmoid)", color='red', style='dashed')
    output = 'Y (Sigmoid)'
    output_layer.node(output, color='pink', fontcolor='black', width='0.8', height='0.8', fixedsize='true')

# Edges from input to hidden layer 1
for i in inputs:
    for h1 in hidden1:
        nn_graph.edge(i, h1)

# Edges from hidden layer 1 to hidden layer 2
for h1 in hidden1:
    for h2 in hidden2:
        nn_graph.edge(h1, h2)

# Edges from hidden layer 2 to output
for h2 in hidden2:
    nn_graph.edge(h2, output)

# Render the graph
nn_graph.render('deep_neural_network_two_hidden_layers_updated', view=True)
```
