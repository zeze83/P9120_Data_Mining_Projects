# Lab 2  Regression-based/Tree-based/SVM models for classification using sklearn

In this lab, we will implement a series of machine learning methods for classification using sklearn package. We will use the built-in cross validation to select the best hyper-parameters for each model, and we will compare their performances using Accuracy/AUC/AUPRC

```python
# Basic Packages
import matplotlib.pyplot as plt # Plotting library
import numpy as np  # Numerical library
import pandas as pd # Dataframe library
```

## Generate data and split train/test sets

```python
import sklearn.datasets
# Generate a dataset and plot it
np.random.seed(0)
X, y = sklearn.datasets.make_moons(200, noise=0.1) # 200 samples, the larger the noise, the less clear the moons are
plt.scatter(X[:,0], X[:,1], s=40, c=y, cmap=plt.cm.Spectral)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# To store the performance of the different models
results = []
```

For each of the model, please:
1. Set up a model
2. Use cross-validation to find the best hyperparameters from the given grid
3. Train the model on the training set
4. Evaluate the model on the test set
5. Store the results

## Regression-based model

### Logistic regression with L1-norm

```python
# Logistic Regression with L1 regularization, use cv to find the best hyperparameter
import sklearn.linear_model # Logistic Regression in sklearn
from sklearn.linear_model import LogisticRegression
from sklearn import metrics # metrics to evaluate the model
from sklearn.model_selection import GridSearchCV # Grid search to find the best hyperparameter

params = {"C": np.logspace(-3,3,7)} # Inverse of regularization strength

# Initialize the model          
model = LogisticRegression(penalty='l1', solver='saga', tol=1e-8) 

# Initialize the GridSearchCV with 5-fold cross-validation
grid = GridSearchCV(model, param_grid=params, cv=5, scoring='roc_auc')

# Fit the model to the training data
grid.fit(X_train, y_train)

# Get the best model
print('Best hyperparameter:', grid.best_params_)
best_model = grid.best_estimator_

# Get the predictions
y_pred = best_model.predict(X_test)
y_pred_prob = best_model.predict_proba(X_test)[:,1]

# Get the performance
test_accuracy = metrics.accuracy_score(y_test, y_pred)
test_auc = metrics.roc_auc_score(y_test, y_pred_prob)
test_auprc = metrics.average_precision_score(y_test, y_pred_prob)

# Store the results
results.append({"model": 'Logistic regression with L1 penalty', "test_accuracy": test_accuracy, "test_auc": test_auc, "test_auprc": test_auprc})
pd.DataFrame(results)
```

## Tree-based model

### Decision Tree

```python
# decision tree for classification
from sklearn.tree import DecisionTreeClassifier
# ?DecisionTreeClassifier
params = {"max_depth": np.arange(1, 11)}

### start your code here 
# Grid search
model = DecisionTreeClassifier() 
# criterion='gini', max_depth=params, random_state=42
grid = GridSearchCV(model, param_grid=params, cv=5, scoring='roc_auc')

# Get the best model
grid.fit(X_train, y_train)
best_model = grid.best_estimator_

# Get the predictions
y_pred = best_model.predict(X_test)
y_pred_prob = best_model.predict_proba(X_test)[:,1]

# Get the performance, test_accuracy, test_auc, test_auprc
test_accuracy = metrics.accuracy_score(y_test, y_pred)
test_auc_dt = metrics.roc_auc_score(y_test, y_pred_prob)
test_auprc_dt = metrics.average_precision_score(y_test, y_pred_prob)

### end your code here

# Store the results
results.append({"model": 'Decision tree', "test_accuracy": test_accuracy, "test_auc": test_auc, "test_auprc": test_auprc})
pd.DataFrame(results)
```

### Random Forest

```python
# Random Forest for classification
from sklearn.ensemble import RandomForestClassifier
params = {"n_estimators": [100, 200, 300, 400, 500],"max_depth": np.arange(1,11)} # more parameters can be added like "max_depth", "min_samples_split", "min_samples_leaf", "max_features"

### start your code here 
# Grid search
model = RandomForestClassifier() 
grid = GridSearchCV(model, param_grid=params, cv=5, scoring='roc_auc')

# Get the best model
grid.fit(X_train, y_train)
print('Best hyperparameter:', grid.best_params_)
best_model = grid.best_estimator_

# Get the predictions
y_pred = best_model.predict(X_test)
y_pred_prob = best_model.predict_proba(X_test)[:,1]

# Get the performance, test_accuracy, test_auc, test_auprc
test_accuracy = metrics.accuracy_score(y_test, y_pred)
test_auc = metrics.roc_auc_score(y_test, y_pred_prob)
test_auprc = metrics.average_precision_score(y_test, y_pred_prob)

### end your code here

# Store the results
results.append({"model": 'Random forest', "test_accuracy": test_accuracy, "test_auc": test_auc, "test_auprc": test_auprc})
pd.DataFrame(results)
```

### Gradient Boosting

```python
# Gradient Boosting for classification
from sklearn.ensemble import GradientBoostingClassifier
params = {"n_estimators": [100, 200, 300, 400, 500], "learning_rate": np.logspace(-3,0,4),"max_depth": np.arange(1,11)} # more parameters can be added like "max_depth", "min_samples_split", "min_samples_leaf", "max_features"
### start your code here 
# Grid search
model = GradientBoostingClassifier()
grid = GridSearchCV(model, param_grid=params, cv=5, scoring='roc_auc')

# Get the best model
grid.fit(X_train, y_train)
print('Best hyperparameter:', grid.best_params_)
best_model = grid.best_estimator_

# Get the predictions
y_pred = best_model.predict(X_test)
y_pred_prob = best_model.predict_proba(X_test)[:,1]

# Get the performance, test_accuracy, test_auc, test_auprc
test_accuracy = metrics.accuracy_score(y_test, y_pred)
test_auc = metrics.roc_auc_score(y_test, y_pred_prob)
test_auprc = metrics.average_precision_score(y_test, y_pred_prob)

### end your code here

# Store the results
results.append({"model": 'Gradient boosting', "test_accuracy": test_accuracy, "test_auc": test_auc, "test_auprc": test_auprc})
pd.DataFrame(results)
```

## SVM

```python
# Support Vector Machine for classification
from sklearn.svm import SVC
params = {"C": np.logspace(-3,3,7), "gamma": np.logspace(-3,3,7)} # more parameters can be added like "kernel", "degree", "coef0"

### start your code here 
# Grid search
model = SVC(probability = True, class_weight='balanced')
grid = GridSearchCV(model, param_grid=params, cv=5, scoring='roc_auc')

# Get the best model
grid.fit(X_train, y_train)
print('Best hyperparameter:', grid.best_params_)
best_model = grid.best_estimator_

# Get the predictions
y_pred = best_model.predict(X_test)
y_pred_prob = best_model.predict_proba(X_test)[:,1]

# Get the performance, test_accuracy, test_auc, test_auprc
test_accuracy = metrics.accuracy_score(y_test, y_pred)
test_auc = metrics.roc_auc_score(y_test, y_pred_prob)
test_auprc = metrics.average_precision_score(y_test, y_pred_prob)

### end your code here

# Store the results
results.append({"model": 'Support vector machine', "test_accuracy": test_accuracy, "test_auc": test_auc, "test_auprc": test_auprc})
pd.DataFrame(results)
```

## Neural Network

```python
# Neural Network for classification
from sklearn.neural_network import MLPClassifier
params = {"hidden_layer_sizes": [(4,), (8,)], "alpha": np.logspace(-3,0,4), 'learning_rate_init': np.logspace(-3,0,4)} 

### start your code here 
# Grid search
model = MLPClassifier(max_iter=2000, random_state=42, activation='tanh') 
grid = GridSearchCV(model, param_grid=params, cv=5, scoring='roc_auc')

# Get the best model
grid.fit(X_train, y_train)
print('Best hyperparameter:', grid.best_params_)
best_model = grid.best_estimator_

# Get the predictions
y_pred = best_model.predict(X_test)
y_pred_prob = best_model.predict_proba(X_test)[:,1]

# Get the performance, test_accuracy, test_auc, test_auprc
test_accuracy = metrics.accuracy_score(y_test, y_pred)
test_auc = metrics.roc_auc_score(y_test, y_pred_prob)
test_auprc = metrics.average_precision_score(y_test, y_pred_prob)

### end your code here

# Store the results
results.append({"model": 'Neural network', "test_accuracy": test_accuracy, "test_auc": test_auc, "test_auprc": test_auprc})
pd.DataFrame(results)
```
