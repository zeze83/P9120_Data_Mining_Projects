### (a) Suppose Rs = +1. Consider an equiprobable random policy $\pi$ (i.e., all actions equally likely). Implement policy evaluation algorithm to compute $V \pi$. Report the value function on a grid, similar to the example shown on slide 26 of lecture 10.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
grid_size = (4, 4)  # 4x4 grid
num_states = grid_size[0] * grid_size[1]  # Total states (16 states)
Ry = -5    # Reward for yellow death state (state 0)
Rb = +5    # Reward for blue target state (state 15)
Rs = 1     # Reward for states 1-14
gamma = 1  # Discount factor
actions = ['up', 'down', 'left', 'right']  # Possible actions

# State transitions
def get_next_state(state, action):
    """Get the next state from the given state and action."""
    row, col = divmod(state, grid_size[1])
    if action == 'up':
        next_row = max(0, row - 1)
        next_state = next_row * grid_size[1] + col
    elif action == 'down':
        next_row = min(grid_size[0] - 1, row + 1)
        next_state = next_row * grid_size[1] + col
    elif action == 'left':
        next_col = max(0, col - 1)
        next_state = row * grid_size[1] + next_col
    elif action == 'right':
        next_col = min(grid_size[1] - 1, col + 1)
        next_state = row * grid_size[1] + next_col
    else:
        next_state = state  # No movement if action is invalid
    return next_state

# Policy evaluation algorithm
def policy_evaluation(random_policy=True, theta=1e-4):
    # Initialize value function
    V = np.zeros(num_states)
    delta = float('inf')
    while delta > theta:
        delta = 0
        for state in range(num_states):
            if state == 0:  # Death cell
                v = Ry  # Terminal state value
            elif state == num_states - 1:  # Target cell
                v = Rb  # Terminal state value
            else:
                v = 0
                for action in actions:
                    next_state = get_next_state(state, action)
                    reward = Rs if next_state not in [0, num_states - 1] else Ry if next_state == 0 else Rb
                    v += 0.25 * (reward + gamma * V[next_state])  # Average over all actions
            delta = max(delta, abs(v - V[state]))
            V[state] = v
    return V

V = policy_evaluation()
V_grid = V.reshape(grid_size)
print(V_grid)
```

### (c) Suppose Rs can take values −1, 0 or 1. Which of the three values would cause the optimal policy to return the shortest path to the blue target cell?

### Derive the optimal policy and optimal value function analytically. Draw the optimal policy on the grid and represent the optimal value function on a separate grid, similar to the example shown on slide 14 of lecture 11

```python
import numpy as np

# Define the gridworld parameters
grid_size = (4, 4)  # 4x4 grid
num_states = grid_size[0] * grid_size[1]  # Total states (16 states)
Ry = -5    # Reward for yellow death state (state 0)
Rb = +5    # Reward for blue target state (state 15)
Rs = 1     # Reward for states 1-14
discount_factor = 1.0  # Gamma
actions = ['up', 'down', 'left', 'right']
action_prob = 1 / len(actions)  # Equiprobable random policy

def get_next_state(state, action):
    """Get the next state from the given state and action."""
    row, col = divmod(state, grid_size[1])
    if action == 'up':
        row = max(0, row - 1)
    elif action == 'down':
        row = min(grid_size[0] - 1, row + 1)
    elif action == 'left':
        col = max(0, col - 1)
    elif action == 'right':
        col = min(grid_size[1] - 1, col + 1)
    return row * grid_size[1] + col

def get_rewards(Rs):
    """Return the reward for each state in the grid."""
    rewards = np.full(num_states, Rs)  # Set Rs for all states
    rewards[0] = Ry  # Death state reward
    rewards[num_states - 1] = Rb  # Target state reward
    return rewards

def value_iteration(states, rewards, discount_factor, threshold=1e-4):
    """Perform value iteration to compute the optimal value function and policy."""
    value_function = np.zeros(states)
    policy = np.zeros(states, dtype=int)  # Store optimal actions
    while True:
        delta = 0
        new_value_function = np.zeros(states)
        for state in range(states):
            if state == 0 or state == 15:  # Terminal states
                new_value_function[state] = rewards[state]
                continue

            best_value = float('-inf')
            best_action = None
            for action_idx, action in enumerate(actions):
                next_state = get_next_state(state, action)
                value = rewards[state] + discount_factor * value_function[next_state]
                if value > best_value:
                    best_value = value
                    best_action = action_idx

            new_value_function[state] = best_value
            policy[state] = best_action
            delta = max(delta, abs(best_value - value_function[state]))

        value_function = new_value_function
        if delta < threshold:
            break
    return value_function, policy
```

```python
import gymnasium as gym
import numpy as np
import matplotlib.pyplot as plt

class Agent:
    def __init__(self, policy='eps_greedy', alpha=0.1, epsilon=1, bins=None):
        self.total_reward = 0
        self.policy = policy
        self.alpha = alpha  # learning rate
        self.epsilon = epsilon  # epsilon-greedy
        self.gamma = 1  # discount factor, DO NOT CHANGE
        self.bins = bins if bins else (np.array([0]), np.array([1e20]), np.array([-0.2, 0, 0.2]), np.array([-0.3, 0.3]))
        self.theta = np.zeros([len(self.bins[0]) + 1, len(self.bins[1]) + 1, len(self.bins[2]) + 1, len(self.bins[3]) + 1, 2])
        
    def obs_index(self, state):
        ind = np.zeros(4).astype(int)
        for i in range(len(state)):
            ind[i] = np.digitize(state[i], self.bins[i])
        return tuple(ind)
    
    def q(self, state):
        ind = self.obs_index(state)
        return self.theta[ind]
    
    def q_update(self, last_state, action, reward, state):
        ind = self.obs_index(state)
        ind_last = self.obs_index(last_state)
        delta = (reward + self.gamma * np.max(self.theta[ind]) - self.theta[ind_last + (action,)])
        self.theta[ind_last + (action,)] += self.alpha * delta

    def choose_action(self, state, episode):
        if np.random.rand() > self.epsilon:
            return np.argmax(self.q(state))
        else:
            return np.random.choice([0, 1])

    def gather_reward(self, reward, t):
        self.total_reward += (self.gamma ** t) * reward

    def get_total_reward(self):
        return self.total_reward

    def set_total_reward(self, new_total):
        self.total_reward = new_total


def train_q_learning(env, alpha=0.1, epsilon=1, epsilon_decay=0.995, alpha_decay=0.99, bins=None, episodes=500, max_steps=200):
    agent = Agent(policy='eps_greedy', alpha=alpha, epsilon=epsilon, bins=bins)
    ep_rewards = []
    for episode in range(episodes):
        state, info = env.reset()
        agent.set_total_reward(0)
        done = False
        for t in range(max_steps):
            action = agent.choose_action(state, episode)
            next_state, reward, done, truncated, info = env.step(action)
            agent.q_update(state, action, reward, next_state)
            state = next_state
            agent.gather_reward(reward, t)
            if done or t == max_steps - 1:
                ep_rewards.append(agent.get_total_reward())
                break
        agent.epsilon = max(0.01, agent.epsilon * epsilon_decay)
        agent.alpha = max(0.01, agent.alpha * alpha_decay)
    return agent, ep_rewards


def test_q_learning(env, agent, episodes=50, max_steps=200):
    ep_rewards = []
    for episode in range(episodes):
        state, info = env.reset()
        agent.set_total_reward(0)
        done = False
        for t in range(max_steps):
            action = np.argmax(agent.q(state))
            state, reward, done, truncated, info = env.step(action)
            agent.gather_reward(reward, t)
            if done or t == max_steps - 1:
                ep_rewards.append(agent.get_total_reward())
                break
    return ep_rewards


def plot_rewards(ep_rewards, title="Performance of Estimated Optimal Policy on CartPole"):
    avg_reward = np.round(np.mean(ep_rewards), 1)
    sd_reward = np.round(np.std(ep_rewards), 1)
    plt.figure(figsize=(10, 6))
    plt.plot(ep_rewards, label="Total Reward")
    plt.axhline(y=avg_reward, color='r', linestyle='--', label=f"Mean: {avg_reward}")
    plt.xlabel('Episode')
    plt.ylabel('Total Reward')
    plt.title(title)
    plt.legend()
    plt.show()


# Parameter configurations
param_combinations = [
    {'alpha': 0.1, 'epsilon': 1, 'bins': [np.array([-0.2, 0, 0.2]), np.array([-1, 0, 1]), np.array([-0.2, 0.2]), np.array([-0.3, 0, 0.3])], 'episodes': 500, 'max_steps': 200},
    {'alpha': 0.05, 'epsilon': 0.9, 'bins': [np.array([-0.5, 0, 0.5]), np.array([-2, 0, 2]), np.array([-0.5, 0.5]), np.array([-1, 0, 1])], 'episodes': 1000, 'max_steps': 200},
    {'alpha': 0.01, 'epsilon': 0.8, 'bins': [np.array([-0.2, 0, 0.2]), np.array([-1, 0, 1]), np.array([-0.2, 0.2]), np.array([-0.3, 0, 0.3])], 'episodes': 1500, 'max_steps': 200},
    {'alpha': 0.1, 'epsilon': 0.7, 'bins': [np.array([-0.1, 0.1]), np.array([-0.5, 0.5]), np.array([-0.1, 0.1]), np.array([-0.2, 0.2])], 'episodes': 500, 'max_steps': 100},
]

env = gym.make('CartPole-v1')

for i, params in enumerate(param_combinations):
    print(f"Running configuration {i+1}...")
    agent, training_rewards = train_q_learning(
        env, 
        alpha=params['alpha'], 
        epsilon=params['epsilon'], 
        bins=params['bins'], 
        episodes=params['episodes'], 
        max_steps=params['max_steps']
    )
    plot_rewards(training_rewards, f"Training Rewards for Configuration {i+1}")

    print(f"Testing Configuration {i+1}...")
    test_rewards = test_q_learning(env, agent, episodes=50, max_steps=200)
    plot_rewards(test_rewards, f"Test Rewards for Configuration {i+1}")

env.close()
```

Higher learning rates ( $\alpha$ = 0.1 ) allowed faster updates and better performance.

Lower learning rates ( $\alpha$ = 0.01, 0.05 ) caused slower convergence and poor results.

High initial exploration ( $\epsilon$ = 1.0 ) resulted in low performance during training.

Moderate exploration ( $\epsilon$ = 0.7, 0.8 ) with decay allowed the agent to balance exploration and exploitation, improving performance.

Increasing the number of bins improved the agent’s ability to generalize and learn better policies.

Fewer bins limited the agent’s ability to learn optimal policies.

Longer training did not always lead to better performance when other parameters were poorly set.

In conclusion, we can find that configuration 4 si the bast one since it achieved the highest rewards during both training and test. And it has balanced exploration with sufficient discretization for the state space.
